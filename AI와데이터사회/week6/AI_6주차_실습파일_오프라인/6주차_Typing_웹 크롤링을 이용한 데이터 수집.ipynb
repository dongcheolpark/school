```python
from bs4 import BeautifulSoup    #웹페이지를 파이썬이 이해하기 쉬운구조로 파싱
import urllib.request            # 웹페이지를 요청, 다운로드 등 기능을 가짐
from selenium import webdriver   #브라우저를 제어하는 패키지
import time
```

```python
#크롬 새 창에 해당 주소의 웹페이지가 뜸

# 웹드라이버 지정, 코랩인 경우 아래의 코드 사용
#driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options) 

# 웹드라이버 지정, 주피터 노트북인 경우
driver = webdriver.Chrome('/Users/parkdongcheol/programming/school/AI와데이터사회/week6/AI_6주차_실습파일_오프라인/chromedriver') 

#웹페이지 요청하여 소스코드 가져옴
driver.get("https://search.daum.net/search?w=img&nil_search=btn&DA=NTB&enc=utf8&q=%EC%95%84%EC%9D%B4%EC%9C%A0") # 이미지 검색 URL
time.sleep(5) # 5초 동안 페이지 로딩 기다리며 파이썬은 쉰다.
```

```python
req = driver.page_source
# HTML을 BeautifulSoup이라는 라이브러리를 활용해 검색하기 용이한 상태로 만듦
# soup이라는 변수에 "파싱 용이해진 html"이 담긴 상태가 됨
soup = BeautifulSoup(req, 'html.parser')
soup
```

```python
img = soup.select_one('#imgList > div > a > img')
# 필요한 부분을 추출
print(img)
print(img['src'])
```

```python
img_s = soup.select('#imgList > div > a > img')
img_s
```

```python
i = 1
#어려개의 태그가 저장된 img_s로 부터 하나씩만 img에 가져옴
for img in img_s: 
        #format()함수로 파일명을 문자열로 생성
        #./img/img1.jpg, ./img/img2.jpg  ... 형식으로 저장
        #Colab의 경우
        #savename = '/content/drive/MyDrive/Colab Notebooks/AI_6주차_실습파일_오프라인/img/{}.jpg'.format(i)    
        #주피터노트북의 경우
        savename = './img/{}.jpg'.format(i)            
        urllib.request.urlretrieve(img['src'],savename)        
        i += 1        
    
driver.quit() # 끝나면 닫아주기
```
